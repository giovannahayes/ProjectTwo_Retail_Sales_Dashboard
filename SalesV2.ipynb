```python
#Store CSV into Dataframe for our sales data

dfx = "./Resources_New/train.csv"
train_df = pd.read_csv(dfx,encoding='latin-1')
train_df
```

```python
#lets merge our storeloc data frame with our sales dataframe
inner_join = pd.merge(train_df,  
                      storeloc_df,  
                      on ='Store',  
                      how ='inner') 
inner_join 
```

```python
# new data frame with split value columns 
citystate_df = inner_join["City"].str.split(", ", n = 1, expand = True) 
  
# making separate first name column from new data frame 
inner_join["Cities"]= citystate_df[0] 
  
# making separate last name column from new data frame 
inner_join["States"]= citystate_df[1] 
  
# Dropping old Name columns 
inner_join.drop(columns =["City"], inplace = True) 
  
# df display 
citystate_df

```

```python
# new data frame with split value columns 
newstoreloc_df = storeloc_df["City"].str.split(", ", n = 1, expand = True) 
  
# making separate first name column from new data frame 
storeloc_df["Cities"]= newstoreloc_df[0] 
  
# making separate last name column from new data frame 
storeloc_df["States"]= newstoreloc_df[1] 
  
# Dropping old Name columns 
storeloc_df.drop(columns =["City"], inplace = True) 
  
# df display 
newstoreloc_df
```

```python
newstoreloc_df.reset_index(drop= True, inplace = True)
```

```python
newstoreloc_df.rename(columns={0: 'city'},
          inplace=True)
```

```python
newstoreloc_df.rename(columns={1: 'state'},
          inplace=True)
newstoreloc_df
```

```python
#lets tack on our cities into a list
newcitylist = newstoreloc_df['city'].tolist()
newcitylist
```

```python
latlist = []
longlist = []

for x in range(45):
    city = newcitylist[x]
    geolocator = Nominatim(user_agent="my_user_agent")
    country ="USA"
    loc = geolocator.geocode(city+','+ country)
    latlist.append(loc.latitude)
    longlist.append(loc.longitude)
    
print(latlist)
print(longlist)


```

```python
storeloc_df['Latitude'] = latlist
storeloc_df['Longitude'] = longlist
storeloc_df['Latitude'] = storeloc_df['Latitude'].round(2)
storeloc_df['Longitude'] = storeloc_df['Longitude'].round(2)
storeloc_df
```

```python
#lets merge our storeloc data frame again with our sales dataframe
final_df = pd.merge(inner_join,  
                      storeloc_df,  
                      on ='Store',  
                      how ='inner') 
final_df
```

```python
#remove duplicate columns and rename some columns
final_df.drop(['Cities_x','States_x'], axis=1, inplace=True)

final_df
```

```python
final_df.rename(columns= {'IsHoliday': 'Holiday', 'Cities_y': 'City','States_y': 'State'})
final_df
```
